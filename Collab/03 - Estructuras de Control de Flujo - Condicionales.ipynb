#**Estructuras Condicionales**

Las estructuras de control permiten ejecutar ciertas acciones basadas en una condici√≥n booleana. La idea fundamental es que el programa pueda tomar decisiones y realizar diferentes acciones dependiendo de si una condici√≥n espec√≠fica es verdadera o falsa.






##**Sentencia Condicional " Si "**

La estructura de control if permite que un programa ejecute unas instrucciones cuando se cumplan una condici√≥n. En ingl√©s "if" significa "si" (condici√≥n).

Los condicionales nos permiten tener la capacidad de comprobar condiciones y cambiar el comportamiento del programa de acuerdo a ellas.

###**Condicional con Expresi√≥n (Condici√≥n)**

La sentencia "if" eval√∫a una expresi√≥n (cualquiera que genere un valor booleano). Si la expresi√≥n se eval√∫a True, se realizan las instrucciones dentro del condicional, sino se sale sin hacer nada y contin√∫a con la ejecuci√≥n principal.

Su estructura es la siguiente:
```
if condici√≥n:
  bloque_1
```



````python
num_1 = int(input("Escriba cualquier n√∫mero: "))
num_2 = 25

if num_1 < num_2:  # Aqu√≠ la expresi√≥n es "num_1 < num_2" la cual puede evaluarse como falso o verdadero
  print(f"El n√∫mero {num_1} es menor que {num_2}")

print("Fin de la ejecuci√≥n.")
````

###**Condicional con Valor Booleano**

Aqu√≠ no hay una expresi√≥n como tal, s√≥lo un valor booleano mediante una variable.

Si la variable no es booleana sino de tipo string, lo que se hace es que se evalua si la variable tiene alg√∫n contenido o no, si lo tiene, se considera True, si est√° vac√≠a se considera un False.
Si la variable es num√©rica, se considera True si tiene cualquier valor diferente a 0, y se considera False si es 0.

````python
dato = input("Escriba cualquier dato: ")

if dato:   # Aqu√≠ dato es una variable string
  print(f"Ha introducido el dato {dato}")

print("Fin de la ejecuci√≥n.")
````

````python
# Esto es equivalente a lo anterior

dato = input("Escriba cualquier dato: ")

if dato != "":
    print(f"Ha introducido el dato {dato}")

print("Fin de la ejecuci√≥n.")
````

````python
num = int(input("Escriba cualquier n√∫mero: "))

if num:   # Aqu√≠ num es una variable entera, pero tambi√©n aplica a decimal (float)
    print(f"Ha introducido el n√∫mero {num}")

print("Fin de la ejecuci√≥n.")
````

````python
# Esto es equivalente a lo anterior

num = int(input("Escriba cualquier n√∫mero: "))

if num != 0:
    print(f"Ha introducido el n√∫mero {num}")

print("Fin de la ejecuci√≥n.")
````

###**Condicional con Bifurcaci√≥n ' if... else... '**

La estructura de control "if... else..." permite que un programa ejecute unas instrucciones cuando se cumple una condici√≥n y otras instrucciones cuando no se cumple esa condici√≥n.

Su estructura es la siguiente:
```
if condici√≥n_1:
    bloque 1
else:
    bloque 2
```

````python
num_1 = int(input("Escriba un n√∫mero: "))
num_2 = 30

if num_1 <= num_2:  # Aqu√≠ la expresi√≥n es "num_1 < num_2"
  print(f"El n√∫mero {num_1} es menor o igual que {num_2}")
else:
  print(f"El n√∫mero {num_1} es mayor que {num_2}")

print("Fin de la ejecuci√≥n.")
````

````python
usuario = input("Escriba su nombre de usuario: ")
contrasena = input("Escriba su contrase√±a: ")

if usuario == "usu1" and contrasena == "*123":  # Aqu√≠ la expresi√≥n es compuesta con operador l√≥gico Y
  print(f"Acceso permitido... Bienvenido")
else:
  print(f"Su usuario o contrase√±a no coinciden. Acceso denegado.")
````

````python
cortesia = input("Escriba 'hola' o 'adios': ")

if cortesia == "hola":
    print(f"Hola! C√≥mo est√°s?")
else:
    print("Nos vemos! Que est√©s muy bien...")

print("Fin de la ejecuci√≥n.")
````

###**Condicionales Anidados**

La estructura de control "if... elif... else...", permite evaluar cuando se presenta m√°s de dos posibilidades y es necesario tener m√°s bifurcaciones. En este caso, se evalu√°n exactamente las expresiones dadas en ***if*** y ***elif***, y cualquier otro caso ser√° tratato en el ***else***.

La estructura para este condicional ser√≠a la siguiente:

```
if condici√≥n_1:
    bloque 1
elif condici√≥n_2:
    bloque 2
else:
    bloque 3
```
Esta estructura es equivalente a la siguiente:

```
if condici√≥n_1:
    bloque 1
else:
    if condici√≥n_2:
        bloque 2
    else:
        bloque 3
```

Con esta misma estructura se pueden anidar las estructuras que sean necesarias.

````python
a = int(input("Digite un n√∫mero: "))
b = int(input("Digite otro n√∫mero: "))

if a < b:
   print(f"El n√∫mero a = {a} es menor que el n√∫mero b = {b}")
elif a > b:
   print(f"El n√∫mero b = {b} es menor que el n√∫mero a = {a}")
else:
   suma = a + b
   print(suma)

# Lo que es igual a:

if a < b:
   print(f"El n√∫mero a = {a} es menor que el n√∫mero b = {b}")
else:
   if a > b:
     print(f"El n√∫mero b = {b} es menor que el n√∫mero a = {a}")
   else:
     suma = a + b
     print(suma)
````

###**Condicional Ternario**

El operador ternario es una manera compacta para escribir un condicional √∫nicamente cuando se tienen 2 posibles opciones y cada opci√≥n lleva s√≥lo a una instrucci√≥n.

Su estructura es la siguiente:

`resultado = resultado_si_verdadero if condici√≥n else resultado_si_falso`

N√≥tese que esta forma requiere de una variable en la cual almacenar el resultado del condicional.

Esta forma es equivalente a decir:
```
if condici√≥n:
    resultado_si_verdadero
else:
    resultado_si_falso
```

````python
a = int(input("Digite un n√∫mero: "))

resultado = "Par" if a % 2 == 0 else "Impar"

print(f"El n√∫mero {a} es {resultado}")
````

````python
a = int(input("Digite un n√∫mero: "))

if a % 2 == 0:
  resultado = "Par"
else:
  resultado = "Impar"
````

````python
a = int(input("Digite un n√∫mero: "))
b = int(input("Digite otro n√∫mero diferente al primero: "))

menor = a if a < b else b

print(f"El n√∫mero menor es {menor}")
````

##**Sentencia Condicional ' en caso de '**

Esta sentencia es utilizada cuando se tienen m√∫ltiples opciones. Tambi√©n tiene la ventaja de que soporta patrones complejos (ej. tuplas, tipos, estructuras anidadas).

Para esta se utiliza la estructura del "match... case..." de la siguiente manera:

```
match variable:
  case valor_1_variable:
    bloque_1
  case valor_2_variable:
    bloque_2
  .
  .
  .
  case _:   # En cualquier otro caso
    bloque_3
```
Tambi√©n puedo recibir varias variables y revisarlas en el case:
```
match (variable1, variable_2):
  case (valor_1_variable_1, valor_1_variable_2):
    bloque_1
  case (valor_2_variable_1, valor_2_variable_2):
    bloque_2
  .
  .
  .   
  case _:   # En cualquier otro caso
    bloque_3
```

````python
dia = int(input("Por favor ingresa un valor del 1 al 7 para cada d√≠a de la semana: "))

match dia:
  case 1:
    print("lunes")
  case 2:
    print("martes")
  case 3:
    print("mi√©rcoles")
  case 4:
    print("jueves")
  case 5:
    print("viernes")
  case 6:
    print("s√°bado")
  case 7:
    print("domingo")
  case _:
    print("Valor no v√°lido")
````

````python
apellido = input("Por favor ingresa el apellido del estudiante que deseas consultar: ")
nombre = input("Por favor ingresa el nombre del estudiante que deseas consultar: ")

apellido = apellido.capitalize() # Recordemos que capitalize() Coloca la primera letra en may√∫scula y las dem√°s en min√∫scula
nombre = nombre.capitalize()     # Aqu√≠ garantizamos que no importa c√≥mo lo ingrese el usuario, siempre estar√° igual que en el 'case'

match (apellido, nombre):
  case ("Alvarez","Julian"):
    print("Promedio del Semestre: 4.5")
  case ("Gomez","Natalia"):
    print("Promedio del Semestre: 4.8")
  case ("Gomez","Andres"):
    print("Promedio del Semestre: 3.6")
  case ("Rodriguez","Catalina"):
    print("Promedio del Semestre: 4.1")
  case _:
    print("Estudiante no encontrado")
````
