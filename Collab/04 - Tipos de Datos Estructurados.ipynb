#**Datos Estructurados**

Estos son considerados "datos no primitivos" o "datos compuestos" porque se construyen combinando o agrupando varios valores de tipos de datos primitivos o incluso otros tipos de datos compuestos.

En el caso de Python, la distinción entre primitivos y no primitivos no es tan rígida como en algunos otros lenguajes, ya que todos estos tipos de datos son objetos y tienen sus propios métodos asociados.

##**Listas o list**

Una **lista** es una estructura de datos que permite almacenar y organizar elementos de manera ordenada y mutable (que pueden modificarse sus elementos). Se define utilizando **corchetes** **[ ]** y los elementos dentro de la lista están separados por comas.

Cada elemento en la lista puede ser de cualquier tipo de datos, incluyendo otros **objetos**, como listas anidadas.

Las listas al ser una estructura de secuencias ordenadas tienen indices que permiten acceder a un dato en una posición en específica. Los números en rojo son las posiciones o índices, así:

image.png

Puedo **leer** o **modificar** datos en una **lista** mediante un ciclo **for** haciendo lecturas y asignaciones directas: mi_lista[índice] = n.

```python
mi_lista = [1, 2, 3, 'cuatro', 5.0]
```

```python
# Leer
print(mi_lista[2])
```

```python
# Modificar
mi_lista[3] = 4
print(mi_lista)
```

Para realizar otras acciones debo utilizar las operaciones propias de las listas:
* .append(elemento) → Añade un elemento al final.
* .extend(iterable) → Añade múltiples elementos.
* .remove(elemento) → Elimina la primera ocurrencia del elemento. Genera error si no existe.
* .clear() → Vacía la lista.

```python
lista = [1, 2, 3]
print(f"Original: {lista}")
```

```python
print(lista)
```

```python
lista.append(4)
print(f"Con append 4: {lista}")
```

```python
lista.extend([5, 6, 2])
print(f"Con extend [5, 6, 2]: {lista}")
```

```python
lista.remove(2)
print(f"Con remove 2: {lista}")
```

```python
lista.clear()
print(f"Con clear: {lista}")
```

##**Tuplas o tuple**

Las **tuplas** son secuencias ordenadas de valores, separados por comas y encerrados entre **paréntesis** **()**. Estos tipos de datos son inmutables, es decir que una vez declarados ya no se pueden modificar sus elementos.

Puedo **leer** una **tupla** mediante un ciclo **for**, pero no puedo escribir en ella. Si requiero modificarla debo convertirla en otro tipo de dato estructurado.

```python
mi_tupla = (1, 2, 3, 'cuatro', 5.0)
```

```python
print(mi_tupla[2])
```

```python
print(mi_tupla)
```

##**Conjuntos o set**

Los **conjuntos** son colecciones de elementos no ordenados, mutables y sin elementos duplicados, encerrados entre **llaves {}**.

Con los conjuntos se pueden realizar operaciones de comparación y de álgebra de conjuntos, que son propias de este tipo de datos. Suponiendo dos conjuntos A y B:
* Unión '|' : elementos en A o B o en ambos
* Intersección '&' : elementos en A y B
* Diferencia '-' : elementos en A pero no en B
* Diferencia simétrica '^' : elementos en A o B, pero no en ambos

Un conjunto vacio es representado por set().

En vez de los operadores también se pueden utilizar las funciones propias de los conjuntos para cada operación:
* .union(conjunto)
* .intersection(conjunto)
* .difference(conjunto)
* .symmetric_difference(conjunto)


```python
conjunto1 = {1, 2, 3}
conjunto2 = {3, 4, 5}
```

```python
union1 = conjunto1.union(conjunto2)
print(f"Unión función: {union1}")

union2 = conjunto1 | conjunto2
print(f"Unión operador: {union2}")
```

```python
interseccion1 = conjunto1.intersection(conjunto2)
print(f"Interseccion función: {interseccion1}")

interseccion2 = conjunto1 & conjunto2
print(f"Interseccion operador: {interseccion2}")
```

```python
diferencia1 = conjunto1.difference(conjunto2)
print(f"Diferencia función: {diferencia1}")

diferencia2 = conjunto2 - conjunto1
print(f"Diferencia operador: {diferencia2}")
```

```python
diferenciasim1 = conjunto1.symmetric_difference(conjunto2)
print(f"Diferencia Simétrica función: {diferenciasim1}")

diferenciasim2 = conjunto2 ^ conjunto1
print(f"Diferencia Simétrica operador: {diferenciasim2}")
```

Para modificar un conjunto se cuenta con las siguientes funciones:
* .add(elemento) : Añadir un elemento
* .update(iterable) : Añade múltiples elementos
* .remove(elemento) : Eliminar un elemento (error si no existe)
* .discard(elemento) : Eliminar si existe (sin error)
* .clear()  : Vaciar el conjunto

```python
mi_set = {1, 2, 3}
print(f"Original: {mi_set}")
```

```python
mi_set.add(4)
print(f"Con add 4: {mi_set}")
```

```python
mi_set.update([4, 5])
print(f"Con update [4, 5]: {mi_set}")
```

```python
mi_set.remove(2)
print(f"Con remove 2: {mi_set}")
```

```python
mi_set.discard(99)
print(f"Con discard 99: {mi_set}")
```

```python
mi_set.clear()
print(f"Con clear: {mi_set}")
```

##**Diccionarios o dict**

Un **diccionario** es un similar a una lista, pero su composición esta conformada con elementos **clave (key)** : **valor (value)**. Se representa mediante **llaves {}** pero sus elementos tiene la forma "clave":valor.

En un diccionario la manera de acceder al **valor** (value) es por medio de las **claves** (key).

Puedo **leer** o **modificar** datos en un **diccionario** mediante un ciclo **for** haciendo lecturas y asignaciones directas: diccionario[clave] = n.

Un diccionario vacio es representado por unas llaves vacias {}.

A continuación se puede observar como es la estructura de un diccionario:

```python
diccionario = {"Nombre" : "Pedro", "Celular" : 31313131310}
print(diccionario["Nombre"])
```

```python
diccionario["Nombre"] = "Juan"
print(diccionario["Nombre"])
```

En general un diccionario es como una lista en cuanto a su longitud, ya que esta mutable y dinámica, es decir se pueden agregar cualquier cantidad y tipos de datos.

Las claves o llaves de un diccionario en general pueden ser strings o números enteros y los valores pueden ser cualquier cosa como una lista, una tupla, incluso otro diccionario.



```python
# Diccionario con elementos de tipo diccionario

diccionario = { 15131313 : {"Nombre" : "Pedro", "Celular" : 31313131310},
                16161616 : {"Nombre" : "Carlos", "Celular" : 3126262626},
                21212122 : {"Nombre" : "Juan", "Celular" : 3112626266}
              }
print(diccionario[15131313])
```

Para modificar un diccionario se cuenta con las siguientes funciones:

* .update(otro_dict) → Añade o actualiza pares clave-valor.
* .pop(clave) → Equivalente a remove (elimina y devuelve el valor, error si no existe).
* .pop(clave, default) → Evita errores (como discard).
* .clear() → Vacía el diccionario.

```python
mi_dicc = {"a": 1, "b": 2}
print(f"Original: {mi_dicc}")
```

```python
mi_dicc["c"] = 3
print(f"Asignar ('c',3): {mi_dicc}")
```

```python
mi_dicc.update({"d": 4})
print(f"Con update ('d',4): {mi_dicc}")
```

```python
mi_dicc.pop("b")
print(f"Con pop b: {mi_dicc}")
```

```python
mi_dicc.pop("z", None)
print(f"Con pop z: {mi_dicc}")
```

```python
mi_dicc.clear()
print(f"Con clear: {mi_dicc}")
```

##**Conversión de Tipos**

```python
tupla = (1, 2, 3)
print(tupla)
```

```python
lista = list(tupla)
print(lista)
```

```python
conjunto = set(tupla)
print(conjunto)
```

```python
tupla = tuple(lista)
print(tupla)
```

```python
lista = [["a", 3], [2,5]]

diccionario = dict(lista)
print(diccionario)
```

```python
mi_dicc = {"a": 1, "b": 2}
lista1 = list(mi_dicc)
print(lista1)
```

```python
lista2 = mi_dicc.items()
print(lista2)
```
