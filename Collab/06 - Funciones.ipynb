##**Funciones**

Cuando se deﬁne una función, se especiﬁca el nombre y la secuencia de sentencias. Más adelante, se puede “llamar” a la función por ese nombre.

####Ejemplo:

La manera de hacer un llamado a una función es nombre de la función, apertura de parentesis, args (si la función los requiere) y el cierre de parentesis.


* print(args)
* input(args)
* range(args)

###**Definición de Funciones Nuevas**

Las funciones ayudan a modularizar y organizar el código, esto promueve la reutilización del código, la claridad y la mantenibilidad del mismo.

Cuando se desea definir una nueva función siempre que tenga una tarea o conjunto de tareas que se repitan en su programa y que puedan beneficiarse de la modularidad, la reutilización y la claridad proporcionadas por la abstracción funcional.

####**Ejemplos**

Se desea definir el promedio de 4 numeros los cuales van a ser ingresados por el usuario, la función debe calcular el promedio cada vez que sea utilizada.

```python
# Función sin parámetros y valor de retorno

def calcular_promedio():

  contador = 1
  sumatoria = 0

  while contador <= 4:
      numero = int(input("Digite un número "))#---> Cast de datos
      sumatoria += numero
      contador += 1

  return(sumatoria/4)

# ---- Fin de la función ----


# Llamado a la función

prom = calcular_promedio()

print("El promedio de los datos introducidos es: ", prom)

prom2 = calcular_promedio()

print("El promedio de los datos introducidos es: ", prom2)
```

Del ejercicio anterior modificar para que el software calcule el promedio de los números introducidos, el software termina y hace el calculo si el usuario introduce un 0.

```python
def promedio():

  contador = 1
  sumatoria = 0

  while True:
      numero = int(input("Digite un número ")) #---> Cast de datos
      if numero == 0:
        contador -= 1
        break
      sumatoria += numero
      contador += 1

  return (sumatoria/contador)

# ---- Fin de la función ----

# Llamado a la función

prom = promedio()
print("El promedio de los datos introducidos es: ", prom)
```

```python
# Función con parámetros y valor de retorno

def sumar(a, b):
  c = a + b
  return c

# ---- Fin de la función ----

# Llamado a la función

suma = sumar(2, 6)
print(suma)

print(sumar(8,87))
```

```python
# Función con parámetros y sin valor de retorno

def saludar(nombre, mensaje):
  print(f"{mensaje} {nombre}!")

# ---- Fin de la función ----

# Llamado a la función

saludar("Carlos", "Muy buenos días")
```
