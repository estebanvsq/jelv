## **Listas (Mutables y Ordenadas)**

For + Listas:

Crea una lista de precios. Usando for, aumenta cada precio en 10% y muestra los nuevos precios.

```python
precios = [9500, 6550, 45000, 86300, 76350]
new_precios = []

for precio in precios:
  new_precios.append(int(precio * 1.1))

print(new_precios)
```

While + Listas:

Pide n√∫meros hasta que el usuario ingrese 0. Almac√©nalos en una lista y muestra la suma.

```python
num = 1
lista = []
suma = 0

while num != 0:
  num = int(input("Por favor ingrese un n√∫mero. Para salir digite 0. \n"))
  if num == 0:
    break
  lista.append(num)
  suma = suma + num

print(f"Lista: {lista}")
print(f"Suma: {suma}")
```

## **Funciones**

Funci√≥n para calcular el factorial:

Crea una funci√≥n que calcule el factorial de un n√∫mero entero positivo usando un ciclo for. (1*2*3*4*5*6*...)

```python
def factorial(f):

  cont = 1
  fact = 1

  while cont <= f:
    fact = fact * cont
    cont +=1

  return fact
#--------

ftrl = int(input("A qu√© n√∫mero desea calcularle el factorial? "))
print(f"{ftrl}! = {factorial(ftrl)}")
```

Funci√≥n para validar pal√≠ndromos:

Crea una funci√≥n que determine si una cadena es un pal√≠ndromo (se lee igual al derecho y al rev√©s).

```python
pal = input("escriba la palabra o la frase que quiere comprobar: ")
pal = pal.lower()

pal = pal.replace(" ", "")
print(pal)

lista_der = []
lista_rev = []

lista_der.extend(pal)
print(lista_der)

l = len(lista_der)

while l > 0:
  lista_rev.append(lista_der[l-1])
  l -= 1

print(f"Lista derecho: {lista_der}")
print(f"Lista reversa: {lista_rev}")

if lista_der == lista_rev:
  print("La palabra es pal√≠ndromo")
else:
  print("La palabra NO es pal√≠ndromo")
```
