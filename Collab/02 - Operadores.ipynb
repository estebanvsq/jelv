# **Operaciones Matemáticas**

Las operaciones directas, ya sean con los datos o con las variables, sólo se usan para mostrar el resultado o usarlo una sola vez en algo en particular.

Si se requiere seguir usando el resultado obtenido, este debe guardarse en una variable y esta se puede seguir utilizando.

###**Suma, Resta y Negación**

````python
# Suma de los datos directos
print(f"9 + 5 = {9 + 5}")

# Suma de la variables directas
a = 9
b = 5
print(f"{a} + {b} = {a + b}")
print(a,"+", b, "=", a + b)

# Suma de la variables guardando el resultado en otra variable
suma = a + b
print(f"{a} + {b} = {suma}")
````

````python
# Resta de los datos directos
print(f"9 - 5 = {9 - 5}")

# Resta de la variables directas
a = 9
b = 5
print(f"{a} - {b} = {a - b}")

# Suma de la variables guardando el resultado en otra variable
resta = a - b
print(f"{a} - {b} = {resta}")
````

````python
# Negación en dato directo
print(f"15 = {-15}")

# Resta de la variables directas
c = 15
print(f"{c} = {-c}")

# Suma de la variables guardando el resultado en otra variable
d = -c
print(f"{c} = {d}")
````

###**Multiplicación, Potenciación (Exponente) y Radicación**

````python
# Multiplicación de los datos directos
print(f"9 * 5 = {9 * 5}")

# Multiplicación de la variables directas
a = 9
b = 5
print(f"{a} * {b} = {a * b}")

# Multiplicación de la variables guardando el resultado en otra variable
multiplicacion = a * b
print(f"{a} * {b} = {multiplicacion}")
````

````python
# Potenciación de los datos directos
print(f"9 ** 5 = {9 ** 5}")

# Potenciación de la variables directas
a = 9
b = 5
print(f"{a} ** {b} = {a ** b}")

# Potenciación de la variables guardando el resultado en otra variable
potenciacion = a ** b
print(f"{a} ** {b} = {potenciacion}")

# Potenciación mediante funciones de la librería "math"
import math    # También se puede llamar como from math import *
math.pow(9, 5)  # 9 a la 5
math.pow(a, b)  # "a" a la "b"

# Potenciación mediante funciones de la librería "math"
from math import pow   # importo sólo el método que interesa de la libreria math
pow(9, 5)   # Llamo al método sin el nombre de la libreria
pow(a, b)
````

````python
# Radicación de los datos directos
print(f"Raíz quinta de 9: 9 ** (1/5) = {9 ** (1/5)}")

# Radicación de la variables directas
a = 9
b = 5
print(f"Raíz quinta de 9: {a} ** {(1/b)} = {a ** (1/b)}")

# Radicación de la variables guardando el resultado en otra variable
radicacion = a ** (1/b)
print(f"Raíz quinta de 9: {a} ** {(1/b)} = {radicacion}")

# Potenciación mediante funciones de la librería "math"
pow(9, 1/5)  # Sacar la raíz quinta de 9
pow(a, 1/b)

math.sqrt(9)  # sqrt() sólo sirve para la raíz cuadrada, raíz cuadrada de 9
math.sqrt(a)
````

###**División, División Entera y Módulo**

````python
# División de los datos directos
print(f"9 / 5 = {9 / 5}")

# División de la variables directas
a = 9
b = 5
print(f"{a} / {b} = {a / b}")

# División de la variables guardando el resultado en otra variable
division = a / b
print(f"{a} / {b} = {division}")
````

````python
# División Entera de los datos directos
print(f"9 // 5 = {9 // 5}")

# División Entera de la variables directas
a = 9
b = 5
print(f"{a} // {b} = {a // b}")

# División Entera de la variables guardando el resultado en otra variable
division_entera = a // b
print(f"{a} // {b} = {division_entera}")
````

````python
# División Módulo de los datos directos
print(f"9 % 5 = {9 % 5}")

# División Módulo de la variables directas
a = 9
b = 5
print(f"{a} % {b} = {a % b}")

# División Módulo de la variables guardando el resultado en otra variable
division_modulo = a % b
print(f"{a} % {b} = {division_modulo}")
````

###**Otras Operaciones**

**Redondeo**

````python
# Redondeo de los datos directos
print(f"Redondear 9.65489 = {round(9.65489)}")
print(f"Redondear 9.65489 a 3 cifras = {round(9.65489, 3)}")

# Redondeo de la variables directas
e = 9.65489
f = 3
print(f"Redondear {e} = {round(e)}")
print(f"Redondear {e} a {f} cifras = {round(e, f)}")

# Redondeo de la variables guardando el resultado en otra variable
redondeo_1 = round(e)
print(f"Redondear {e} = {redondeo_1}")

redondeo_2 = round(e, f)
print(f"Redondear {e} a {f} cifras = {redondeo_2}")
````

**Valor absoluto**

````python
# abs() es un método o función propio de python

# Valor Absoluto de la variable directa
print(f"|{d}| = {abs(d)}")

# Valor Absoluto de la variable guardando el resultado en otra variable
v_absoluto = abs(d)
print(f"|{d}| = {v_absoluto}")
````

**Logaritmos**

````python
# Logaritmo de la variable directa
print(f"Log {a} = {math.log(a)}")

# Logaritmo de la variable guardando el resultado en otra variable
logaritmo = math.log(a)
print(f"Log {a} = {logaritmo}")
````

**Trigonometría Básica**

````python
# Seno y Coseno del dato directo
print(f"Seno 90 = {math.sin(90)}")
print(f"Coseno 90 = {math.cos(90)}")
print(f"Tangente 90 = {math.tan(90)}")

# Seno y Coseno de la variable directa
g = 90
print(f"Seno 90 = {math.sin(g)}")
print(f"Coseno 90 = {math.cos(g)}")
print(f"Tangente 90 = {math.tan(g)}")

# Logaritmo de la variable guardando el resultado en otra variable
seno = math.sin(g)
coseno = math.cos(g)
tangente = math.tan(g)
print(f"Seno 90 = {seno}")
print(f"Coseno 90 = {coseno}")
print(f"Tangente 90 = {tangente}")
````

#**Operaciones con Cadenas**

###**Concatenación**

**Concatenar** es unir dos textos o variables de tipo texto. Para ello se utiliza el símbolo "**+**" pero los operandos son cadenas.

````python
# Concatenar textos directos
print(f"'Hola' + 'Mundo' = {'Hola ' + 'Mundo'}")

# Concatenar las variables directas
h = "Hola "
i = "Mundo"
print(f"{h} + {i} = {h + i}")

# Concatenar las variables guardando el resultado en otra variable
concatenacion = h + i
print(f"{h} + {i} = {concatenacion}")
````

###**Longitud de una Cadena**

La longitud **len**(variable) mide la longitud de la cadena de la variable.

````python
# Longitud de textos directos
print(f"La longitud de 'Hola Mundo' es = {len('Hola Mundo')}")

# Longitud de textos en variables directas
j = "Hola Mundo"
print(f"La longitud de {j} es = {len(j)}")

# Longitud de textos en variables guardando el resultado en otra variable
longitud = len(j)
print(f"La longitud de {j} es = {longitud}")
````

###**Convertir a minúsculas o MAYÚSCULAS**

Existen funciones específicas para convertir una variable de tipo string de tal manera se maneje el comportamiento de las minúsculas y las mayúsculas. Esto se suele utilizar cuando se necesita comparar dos datos en modo texto, para garantizar que sean exactamente iguales, ya sea porque esté todo en minúscula o todo en mayúscula.

image.png

Con estas otras dos funciones verificamos si todos los caracteres son mayúsculas o minúsculas:
```
.isupper() → es mayúscula?
.islower() → es minúscula?
```

````python
texto_1 = "Python"
texto_2 = "python es genial"
texto_3 = "PyTHoN"

# Todo a mayúsculas
texto_1 = texto_1.upper()
print(texto_1)
print(f"El texto {texto_1} está en mayúsculas? {texto_1.isupper()}, o está en minúsculas? {texto_1.islower()}")

# Todo a minúsculas
texto_1 = texto_1.lower()
print(texto_1)
print(f"El texto {texto_1} está en mayúsculas? {texto_1.isupper()}, o está en minúsculas? {texto_1.islower()}")

# Tipo Capital
texto_2 = texto_2.capitalize()
print(texto_2)

# Mayúscula primera letra
texto_2 = texto_2.title()
print(texto_2)

# Invertir mayúsculas y minúsculas
texto_3 = texto_3.swapcase()
print(texto_3)
````

###**Eliminación de Espacios o Caracteres**

Aveces es necesario eliminar espacios en blanco o algún caracter de una cadena. Para ellos se tienen las siguientes funciones:

image.png

````python
texto_4 = " !!Hola¡ !Mundo¡¡ "
texto_4 = texto_4.strip() # Elimina los espacios en blanco
print(texto_4)

texto_4 = "!!Hola¡ !Mundo¡¡"
texto_4 = texto_4.lstrip("!") # Elimina todas las apariciones del caracter indicado al inicio
print(texto_4)

texto_4 = texto_4.rstrip("¡") # Elimina todas las apariciones del caracter indicado al final
print(texto_4)
````

###**Reemplazo y División**

También es posible reemplazar, dividir o unir una cadena en subcadenas. Para ellos utilizamos las siguientes funciones:

image.png

````python
texto_5 = "a, b, c. e, f. e"

texto_5 = texto_5.replace(". e", ", d") # Reemplazar ". e" por ", d" cada que lo encuentre en la cadena
print(texto_5)

texto_5 = texto_5.split(",") # Separar la cadena en una lista de elementos, dividiéndola cada que encuente el separador ","
print(texto_5)

conector = ";"
texto_5 = conector.join(texto_5) # Unir una lista de elementos con el conector ";" en una sola cadena
print(texto_5)
````

###**Eliminación de tildes**

No hay una función especial que me permita quitar tildes o caracteres especiales de un texto, pero con la librería **UNICODEDATA** se puede crear una como esta:

````python
import unicodedata

def remover_tildes(texto):
  # me aseguro de convertirlo en string
  texto = str(texto)

  # Se aplica normalización NFKD (Forma de Normalización de Descomposición de Compatibilidad) al texto:
  # Se toma el texto de entrada y descompone los caracteres en sus componentes básicos y combinadores, y también aplica
  # equivalencia de compatibilidad. Esto significa que caracteres como "á", "é", "ñ", etc., se descomponen en su
  # caracter base (a, e, n) y el caracter combinador (el acento, la tilde). Por ejemplo, "ñ" se convierte en "n" + "~".
  forma_nfkd = unicodedata.normalize('NFKD', texto)

  # '.encode('ascii', 'ignore')': intenta codificar la cadena resultante en ASCII.
  # La parte 'ignore' es crucial porque descarta cualquier caracter que no sea ASCII,
  # que en este caso son los caracteres combinadores (los acentos y tildes).
  # '.decode('utf-8')': decodifica la cadena ASCII en una cadena de Python (que usa UTF-8 por defecto internamente).
  # 'UTF-8' puede representar cualquier caracter en el estándar Unicode. Esto incluye no solo las letras del alfabeto
  # latino, números y símbolos comunes, sino también caracteres de otros idiomas, emojis, símbolos matemáticos y mucho más.
  solo_ascii = forma_nfkd.encode('ascii', 'ignore').decode('utf-8')

  # Retorna el texto convertido sólo en caracteres ASCII sin combinadores
  return solo_ascii

# Ejemplo de uso
texto_con_tildes = "Éste es un ejemplo con tildes y caracteres especiales: ñáéíóúü"
texto_sin_tildes = remover_tildes(texto_con_tildes)
print(f"Texto con tildes: {texto_con_tildes}")
print(f"Texto sin tildes: {texto_sin_tildes}")
````

#**Operaciones Lógicas y de Comparación**

###**Comparación**

Estos operadores comparan dos valores y devuelven True o False:

image.png

````python
# Comparar datos directos
print(f"Es 9 mayor que 4? {9 > 4}")

# Comparar datos en las variables directas
k = 9
l = 4
print(f"Es {k} mayor que {l}? {k > l}")

# Comparar datos en las variables guardando el resultado en otra variable
es_mayor = k > l
print(f"Es {k} mayor que {l}? {es_mayor}")
````

###**Lógicas**

Estos operadores combinan o modifican expresiones booleanas (True/False):

image.png

Estas funcionan de acuerdo con las Tablas de Verdad donde `V->True` y `F->False`:

### Operador AND
| A | B | A and B |
|---|---|--------|
| V | V |   V    |
| V | F |   F    |
| F | V |   F    |
| F | F |   F    |

### Operador OR
| A | B | A or B |
|---|---|-------|
| V | V |   V   |
| V | F |   V   |
| F | V |   V   |
| F | F |   F   |

### Operador NOT
| A | not A |
|---|------|
| V |  F   |
| F |  V   |

````python
# Evaluar expresiones con datos directos
print(f"La expresión '(5 > 3) or (1 == 0)' es = {(5 > 3) or (1 == 0)}")

# Evaluar expresiones con variables directas
m = (5 > 3)
n = (1 == 0)
print(f"La expresión {m} or {n}' es = {m or n}")

# Evaluar expresiones con variables guardando el resultado en otra variable
o_logico = m or n
print(f"La expresión {m} or {n}' es = {o_logico}")
````
