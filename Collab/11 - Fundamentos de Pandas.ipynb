# **Introducción a Pandas**

Pandas es una poderosa librería de Python diseñada para el análisis de datos y la manipulación de datos estructurados. Fue creada por Wes McKinney en 2008 y se ha convertido en una herramienta esencial en el campo de la ciencia de datos y la ingeniería de datos.

## **¿Qué es Pandas?**

Pandas es una biblioteca de software de código abierto para el análisis y la manipulación de datos en Python. Su nombre proviene de "Panel Data" (datos de panel), una referencia a su capacidad para manejar datos estructurados en forma de tablas o DataFrames, similar a una hoja de cálculo de Excel o una base de datos SQL.

Pandas proporciona estructuras de datos de alto rendimiento y fáciles de usar, como Series y DataFrames, que permiten a los usuarios manipular, limpiar, analizar y visualizar datos de manera eficiente.

## **Ventajas de usar Pandas en Análisis de Datos**

*Las ventajas de utilizar Pandas en el análisis de datos incluyen:*

1. Manejo de Datos Estructurados: Pandas proporciona DataFrames, una estructura de datos tabular que permite trabajar con datos estructurados de manera eficiente.

2. Eficiencia en la Manipulación: Pandas es eficiente para operaciones de limpieza y transformación de datos, como filtrado, selección, agrupación y pivotado.

3. Integración con Datos Externos: Permite la importación y exportación de datos desde y hacia varios formatos de archivo, como CSV, Excel, SQL, JSON, y más.

4. Manipulación de Fechas y Tiempos: Ofrece herramientas para trabajar con fechas y horas de manera sencilla.

5. Indexación Personalizada: Permite etiquetar y organizar los datos mediante índices personalizados.

6. Manejo de Datos Faltantes: Pandas facilita la detección y el tratamiento de valores nulos o faltantes en los conjuntos de datos.

7. Funciones Estadísticas y de Agregación: Proporciona herramientas para calcular estadísticas descriptivas y realizar operaciones de agregación en los datos.

8. Operaciones en datos: Permite realizar operaciones de agrupación, combinación, fusión y agregación de datos, y soporta el uso de funciones estadísticas y matemáticas complejas.

## **Estructuras de Datos en Pandas**

Las dos estructuras de datos principales en Pandas son:

1. **Series**: Una Serie es una estructura de datos unidimensional similar a una matriz o una columna en una hoja de cálculo. Cada Serie tiene un índice que permite etiquetar y acceder a los elementos. Puedes pensar en una Serie como una columna de un DataFrame.

2. **DataFrame**: Un DataFrame es una estructura de datos bidimensional que se asemeja a una tabla de base de datos o una hoja de cálculo de Excel. Consiste en filas y columnas, donde cada columna puede contener diferentes tipos de datos. Los DataFrames son especialmente útiles para el análisis de datos tabulares y permite operaciones como filtrar, agregar, pivotar y unir conjuntos de datos.

### **Series: Una Dimensión de Datos Etiquetados**

Una Serie en Pandas es una estructura de datos unidimensional que puede contener datos de diferentes tipos, pero a diferencia de una lista de Python, cada elemento en una Serie está etiquetado con una etiqueta o índice. Estos índices son útiles para acceder y referirse a los datos de manera más intuitiva.

Ejemplo de creación de una Serie en Pandas:

````python
# Creación de una serie a partir de una lista
lista_serie = pd.Series(data=[10,20,30,40,50,60])

print(f"Lista Serie:\n{lista_serie}")
print(type(lista_serie))
````

````python
# Creación de una serie a partir de dos listas
lista_serie = pd.Series(data=[10,20,30,40,50,60], index=["a","b","c","d","e","f"])
print(f"Lista Serie:\n{lista_serie}")
print(type(lista_serie))
````

````python
# Creación de una serie a partir de dos arrays
array_1d = np.array([10,20,30,40,50,60])
indices = np.array(["a","b","c","d","e","f"])
print(f"Array_1d: {array_1d}")
print(f"Índices: {indices}")
print(type(array_1d), type(indices))

array_serie = pd.Series(array_1d, indices)
print(f"\nArray Serie:\n{array_serie}")
print(type(array_serie))
````

````python
# Creación de una serie a partir de array y lista
array_1d = np.array([10,20,30,40,50,60])
indices = ["a","b","c","d","e","f"]
print(f"Array_1d: {array_1d}")
print(f"Índices: {indices}")
print(type(array_1d), type(indices))

array_serie = pd.Series(array_1d, indices)
print(f"\nArray Serie:\n{array_serie}")
print(type(array_serie))
````

### Conversiones a fechas

````python
# Supongamos que se tiene una columna 'fecha' de tipo object en el DataFrame

df = np.array([["24/12/1981 14:30:45"],
               ["24/02/2010 09:10:60"],
               ["25/05/2011 11:00:00"],
               ["27/09/1984 23:15:55"],
               ["31/12/1970 06:26:48"]])

print(f"Array:\n{df}")

titulo=["fecha_hora"]

df = pd.DataFrame(data=df, columns=titulo)
print(f"\nDataframe:\n{df}")
````

````python
print(df['fecha_hora'].dtype)
````

````python
# Se convierte la columna 'fecha' a un objeto de tipo datetime64[ns] que es el tipo estándar de pandas
df['fecha_hora'] = pd.to_datetime(df['fecha_hora'])
print(df)
print(df['fecha_hora'].dtype)
````

````python
# Convierte la columna fecha_hora a string con el formato deseado para visualización
df['fecha_hora_formateada'] = df['fecha_hora'].dt.strftime('%d/%m/%Y %H:%M:%S')
print(df)
print(df['fecha_hora_formateada'].dtype)
````

````python
# para acceder a los componentes de la fecha
df['año'] = df['fecha_hora'].dt.year
df['mes'] = df['fecha_hora'].dt.month
df['dia'] = df['fecha_hora'].dt.day
print(df)
````

````python
df['hora'] = df['fecha_hora'].dt.hour        # Hora (0-23)
df['minuto'] = df['fecha_hora'].dt.minute    # Minuto (0-59)
df['segundo'] = df['fecha_hora'].dt.second   # Segundo (0-59)
df['microsegundo'] = df['fecha_hora'].dt.microsecond  # Microsegundos
print(df)
````

## **Agrupación y Agregación de Datos**

Estructura interna creada (objeto GroupBy de pandas):

Grupos:
  - "Bucaramanga": [R3, R1]
  - "Cúcuta":      [R4]
  - "Cali":        [R5]
  - "Medellín":    [R2]

**Nota 2: Uso de agg()**

>**agg()** permite otras operaciones como **"sum"**, **"mean"**, o funciones personalizadas.

>En este caso, "count" es redundante porque groupby + count se puede simplificar con **value_counts()**.

````python
print(array_df["Edad"].groupby([array_df.Edad]).agg("sum"))
````

````python
print(array_df["Edad"].agg("mean"))
````

## **Limpieza de Datos Nulos**

El método **dropna()** en pandas se utiliza para eliminar filas o columnas con valores faltantes (NaN, None, o NaT) en un DataFrame o Serie.

````python
array_df = array_df.dropna()
````

```
DataFrame.dropna(
    axis=0,          # 0 para filas (default), 1 para columnas
    how='any',       # 'any' (default) o 'all'
    thresh=None,     # Número mínimo de valores no nulos
                     # requeridos para conservar la fila/columna # requeridos
    subset=None,     # Columnas específicas a considerar al
                     # buscar NaN (solo para axis=0)
    inplace=False    # Si es True, modifica el DataFrame original
)
```

Parámetro:

axis
- axis=0 o 'index' (default): Elimina filas con NaN.
- axis=1 o 'columns': Elimina columnas con NaN.

how
- 'any' (default): Elimina si algún valor es NaN.
- 'all': Elimina solo si todos los valores son NaN.

thresh
- Ej: thresh=2 mantiene filas con al menos 2 valores no NaN.

subset

inplace
- False (default): Retorna un nuevo DataFrame.
- True: Modifica el DataFrame original.

***NOTAS***

**NaN vs None:**

>NaN es un valor flotante estándar de NumPy para datos faltantes.

>None es un objeto de Python, pero pandas lo trata como NaN en DataFrames.

**Alternativas a dropna():**

>**fillna()**: Rellena los valores NaN con un valor específico (ej: df.fillna(0)).

>Interpolación: el método **interpolate()** en pandas se utiliza para rellenar valores faltantes (NaN) mediante interpolación, es decir, estimando los valores intermedios basándose en otros puntos de datos. Es especialmente útil para datos secuenciales o series temporales.

```
DataFrame.interpolate(
    method='linear',  # Método de interpolación: 'linear',    
                      #'time', 'polynomial', 'spline','nearest'
    axis=0,           # 0: filas, 1: columnas
    limit=None,       # Número máximo de NaN consecutivos a   
                      # rellenar
    inplace=False,    # Si modifica el DataFrame original
    limit_direction='forward'  # Dirección de relleno:
                               # 'forward', 'backward', 'both'
)
```

¿Cuándo usar interpolate()?
- Datos secuenciales (ej.: series temporales con fechas).
- Cuando los NaN tienen una relación matemática con los datos adyacentes.
- Alternativa a fillna() si los valores no son constantes.

¿Cuándo evitar interpolate()?
- Si los NaN son aleatorios o no tienen patrones.
- En datos categóricos (no numéricos).


**Rendimiento:**

>Usar inplace=True ahorra memoria al no crear un nuevo DataFrame.
