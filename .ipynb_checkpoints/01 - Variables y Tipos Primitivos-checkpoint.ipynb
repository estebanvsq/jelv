# **Tipos Primitivos en Python y Declaraci√≥n de Variables**

Existen varios tipos de datos primitivos en Python, los cuales son:

N√∫meros:
*   Enteros (int) { ...,-3, -2, -1, 0, 1, 2, 3,... } -> Estos valores no tienen l√≠mites topes, m√°s all√° de la capacidad de la memoria de la m√°quina.
*   Decimales (float) { -1.798e+308, ..., -2.3, ..., -1.0, ..., 0.0, ..., 1.5, ..., 2.0, ..., 1.798e+308 } -> Estos valores si tienen l√≠mites establecidos. Y con respecto a la cercan√≠a al CERO, puede llegar hasta 2.225e-308 y -2.225e-308, y soporta una mantisa de entre 15 y 17 cifras decimales.

Texto:
*   Cadena (str)
{ 'ejemplo', "ejemplo", """Este es
una cadena
de bloque""" }

  Con respecto a los l√≠mites de las variables alfanum√©ricas, estas tienen un l√≠mite te√≥rico dado por:
  *  2¬≥¬π - 1 caracteres (~2.1 mil millones) en sistemas de 32 bits.
  *  2‚Å∂¬≥ - 1 caracteres (~9.2 √ó 10¬π‚Å∏) en sistemas de 64 bits ‚Äî aunque esto jam√°s se alcanza por falta de memoria.
  *  O hasta donde soporte la memoria del equipo.


*   Caracter (char) { 'a' }

Booleano (bool) { True - 1, False - 0 }

Las variables tienen dos clasificaciones: cualitativa y cuantitativa. Cuantitativas son las num√©ricas con las que puedo realizar operaciones, y las cualitativas son las cadenas que se expresan de manera textual.

Las variables en Python tienen **tipado din√°mico**, lo que significa que no se declaran con un tipo espec√≠fico, sino que toman el tipo del dato que se le asigne.

"*Asignar una variable*" es almacenar un dato en ella, para lo cual se utiliza el signo "igual" (=).

```python
# Variable tipo Entero (int)

numero_1 = 1100
numero_2 = -1300
numero_3 = 0
```

```python
# Variable tipo Decimal (float)

numero_4 = 12.56
numero_5 = -14.9
numero_6 = 10.0
numero_7 = 0.0
```

```python
# Variable tipo Cadena (str)

cadena_1 = 'Esto es una cadena'
cadena_2 = "Esto tambi√©n es una cadena"
cadena_3 = """Esto es una cadena
de varias l√≠neas"""
cadena_3 = "123"
cadena_4 = "False"
```

```python
# Variable tipo Caracter (char)

caracter_1 = 'a'
caracter_2 = '1'
```

```python
# Variable tipo Booleano (bool)

booleano_1 = True
booleano_2 = False
```

# **Escritura y Lectura por Consola**

Casi siempre ser√° necesario que haya una interacci√≥n del programa con el usuario. Para fines pr√°cticos, esta interacci√≥n se llevar√° a cabo a trav√©s de la **consola**, en la que se pueden escribir o imprimir mensajes a los usuarios, mostrar el contenido de las variables y mostrar el resultado de operaciones, y leer los datos que el usuario puede ingresar.
Para esto utilizaremos las funciones **print()** e **input()**.

```python
# Impresi√≥n de un mensaje

print("Mensaje a imprimir")
```

```python
# Impresi√≥n de una variable

print(numero_1) # Directamente la variable
print(cadena_1) # Directamente la variable
print("N√∫mero entero 2: ", numero_2) # Variable acompa√±ada de un mensaje
print("N√∫mero entero 1: ", numero_1, " N√∫mero entero 2: ", numero_2) # Variable acompa√±ada de un mensaje
print(f"N√∫mero decimal 4: {numero_4}") # Variable acompa√±ada de un mensaje con formato
print(f"N√∫mero decimal 5: {numero_5}   N√∫mero decimal 6: {numero_6}") # Variable acompa√±ada de un mensaje con formato
```

```python
# Impresi√≥n de una operaci√≥n

print(numero_1 + numero_2) # Directamente la operaci√≥n
print("Suma: ", numero_4 + numero_5) # Operaci√≥n acompa√±ada de un mensaje
print(f"Suma: {numero_4 + numero_5}")
print(f"Suma: {numero_4} + {numero_5}")
```

```python
# Lectura de una variable

variable_1 = input()  # Lectura directa de dato sin mensaje a usuario
print(f"Este es el dato ingresado por el usuario en la variable 1: {variable_1}")

variable_2 = input("Por favor ingrese un dato tipo texto: ")  # Lectura de dato con mensaje a usuario
print(f"Este es el dato ingresado por el usuario en la variable 2: {variable_2}")

variable_3 = input("Por favor ingrese un dato tipo num√©rico: ")  # Lectura de dato con mensaje a usuario
print(f"Este es el dato ingresado por el usuario en la variable 3: {variable_3}")

print(f"Por favor ingrese un dato tipo texto: {input()}")
```

# **Conocer el Tipo de Dato de una Variable**

Para conocer el tipo de una variable se utiliza la funci√≥n **type()** con la variable: type(variable). Esta funci√≥n indica el tipo. Normalmente se usa con un print() para poder ver el tipo impreso en la consola.

```python
# Identificar tipos de las variables

print("Tipo variable numero_1: ", type(numero_1))
print("Tipo variable numero_4: ", type(numero_4))
print("Tipo variable cadena_2: ", type(cadena_2))
print("Tipo variable caracter_1: ", type(caracter_1))
print("Tipo variable booleano_1: ", type(booleano_1))
print("Tipo variable variable_2 (ingresada por el usuario): ", type(variable_2))
print("Tipo variable variable_3 (ingresada por el usuario): ", type(variable_3))
```

#**Casting de Datos**

El casting de datos se refiere a la conversi√≥n expl√≠cita de un tipo de datos a otro. Esto se hace para garantizar que los datos se traten de la manera correcta y para permitir operaciones entre diferentes tipos de datos. Para ello se utilizan las siguientes funciones dependiendo en qu√© tipo quiero convertir la variable:

*   **str**(variable)
*   **int**(variable)
*   **float**(variable)
*   **bool**(variable) En este caso, si la variable es una cadena, es False si est√° vac√≠a, y es True si tiene cualquier valor.

Es importante aclarar que todo **input**() recibe una cadena, sin importar si el usuario ingresa un n√∫mero, una palabra o una frase, por lo que el casteo de datos se utiliza obligatoriamente al leer una variable para asegurar que se ingresa el tipo de dato correcto.

```python
# Convertir un n√∫mero en una cadena
cadena_a = str(numero_1) # N√∫mero entero
print(f"N√∫mero 1: {numero_1}, Convertido en Cadena a: {cadena_a}, Nuevo tipo de dato: {type(cadena_a)}")

cadena_b = str(numero_4) # N√∫mero decimal
print(f"N√∫mero 4: {numero_4}, Convertido en Cadena b: {cadena_b}, Nuevo tipo de dato: {type(cadena_b)}")


# Convertir un booleano en una cadena
cadena_c = str(booleano_2) # Booleano
print(f"Booleano 2: {booleano_2}, Convertido en Cadena c: {cadena_c}, Nuevo tipo de dato: {type(cadena_c)}")
```

```python
# Convertir una cadena en un n√∫mero
numero_a = int(cadena_3) # Cadena
print(f"Cadena 3: {cadena_3}, Convertido en N√∫mero a: {numero_a}, Nuevo tipo de dato: {type(numero_a)}")

numero_b = float(cadena_3) # Cadena
print(f"Cadena 3: {cadena_3}, Convertido en N√∫mero b: {numero_b}, Nuevo tipo de dato: {type(numero_b)}")


# Convertir un booleano en un n√∫mero
numero_c = int(booleano_1) # Booleano
print(f"Booleano 1: {booleano_1}, Convertido en N√∫mero c: {numero_c}, Nuevo tipo de dato: {type(numero_c)}")

numero_d = float(booleano_1) # Booleano
print(f"Booleano 1: {booleano_1}, Convertido en N√∫mero d: {numero_d}, Nuevo tipo de dato: {type(numero_d)}")
```

```python
# Convertir un n√∫mero (0/1) a un booleano
booleano_a = bool(numero_3) # N√∫mero entero
print(f"N√∫mero 3: {numero_3}, Convertido en Booleano a: {booleano_a}, Nuevo tipo de dato: {type(booleano_a)}")

booleano_b = bool(numero_7) # N√∫mero decimal
print(f"N√∫mero 7: {numero_7}, Convertido en Booleano b: {booleano_b}, Nuevo tipo de dato: {type(booleano_b)}")


# Convertir un caracter ('0'/'1') a un booleano
booleano_c = bool(caracter_2) # Caracter
print(f"Caracter 2: {caracter_2}, Convertido en Booleano c: {booleano_c}, Nuevo tipo de dato: {type(booleano_c)}")


# Convertir una cadena ('True'/'False') a un booleano
booleano_d = bool(cadena_4) # Cadena
print(f"Cadena 4: {cadena_4}, Convertido en Booleano d: {booleano_d}, Nuevo tipo de dato: {type(booleano_d)}")

booleano_e = bool("") # Cadena vac√≠a
print(f"Cadena vac√≠a: '', Convertido en Booleano e: {booleano_e}, Nuevo tipo de dato: {type(booleano_e)}")
```

```python
# Casteo de datos en el input()

numero_e = int(input("Ingrese un n√∫mero entero: "))
print("Tipo del dato ingresado: ", type(numero_e))

numero_f = float(input("Ingrese un n√∫mero decimal: "))
print("Tipo del dato ingresado: ", type(numero_f))

cadena_d = str(input("Ingrese una cadena: "))
print("Tipo del dato ingresado: ", type(cadena_d))

booleano_f = bool(int(input("Ingrese un booleano (1/0): ")))
print("Tipo del dato ingresado: ", type(booleano_f))
```
